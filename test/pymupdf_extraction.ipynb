```python
import fitz  # PyMuPDF
import pandas as pd

def extract_phrases_with_bounding_boxes(pdf_path, page_number):
    # Open the PDF document
    document = fitz.open(pdf_path)

    # Initialize a list to store phrases and their bounding boxes
    phrase_data = []

    # Loop through each page in the PDF
    for page_num in range(min(page_number,document.page_count)):
        # Get the page
        page = document.load_page(page_num)

        # Extract text with their bounding box information
        text_instances = page.get_text("blocks")

        for instance in text_instances:
            # Each 'instance' is a tuple in the format: (x0, y0, x1, y1, text, block_no, ...)
            x0, y0, x1, y1, text = instance[:5]
            
            # Clean and filter phrases (ignore empty lines)
            if text.strip():
                phrase_data.append({
                    "phrase": text.strip(),
                    "page": page_num + 1,
                    "bounding_box": (x0, y0, x1, y1)
                })

    document.close()
    return phrase_data


```

```python
pdf_file = '/Users/yiminglin/Documents/Codebase/Pdf_reverse/data/raw/certification/VT/Invisible Institue Report.pdf'  # Replace with your PDF file path
file_name = 'Invisible'
approach = 'pypupdf'
page_number = 3


phrases = extract_phrases_with_bounding_boxes(pdf_file,page_number)
out = []
# Display the extracted phrases and bounding boxes
for phrase_info in phrases:
    out.append((phrase_info['page'],phrase_info['phrase'],phrase_info['bounding_box'][0],phrase_info['bounding_box'][1],phrase_info['bounding_box'][2],phrase_info['bounding_box'][3]))

doc_lines_df = pd.DataFrame(out, columns=['Page', 'Phrase', 'x1', 'y1', 'x2', 'y2'])
doc_lines_df.to_csv(file_name + '_' + approach + '.csv')
doc_lines_df
```
